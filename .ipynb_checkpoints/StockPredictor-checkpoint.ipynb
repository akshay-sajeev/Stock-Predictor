```python
df = df.drop(['Date'], axis = 1)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label = f'{stock} Closing Price', linewidth = 2)
plt.title(f'{stock} Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Open'], label = f'{stock} Opening Price', linewidth = 2)
plt.title(f'{stock} Opening Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Opening Price (USD)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['High'], label = f'{stock} High', linewidth = 2)
plt.title(f'{stock} High Over Time')
plt.xlabel('Date')
plt.ylabel('High (USD)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Low'], label = f'{stock} Low', linewidth = 2)
plt.title(f'{stock} Low Over Time')
plt.xlabel('Date')
plt.ylabel('Low (USD)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Volume'], label = f'{stock} Volume', linewidth = 1)
plt.title(f'{stock} Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume (x100,000,000)')
plt.legend()
plt.show()
```

```python
ma100 = df.Close.rolling(100).mean()
```

```python
ma100
```

```python
ma200 = df.Close.rolling(200).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.Close, label = f'{stock} Close Price', linewidth = 1)
plt.plot(ma100, label = f'{stock} Moving Average 100 Price', linewidth = 1)
plt.plot(ma200, label = f'{stock} Moving Average 200 Price', linewidth = 1)
plt.xlabel('Date')
plt.ylabel('High (USD)')
plt.legend()
plt.show()
```

```python
ema100 = df.Close.ewm(span = 100, adjust = False).mean()
```

```python
ema200 = df.Close.ewm(span = 200, adjust = False).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.Close, label = f'{stock} Close Price', linewidth = 1)
plt.plot(ema100, label = f'{stock} Exp. Moving Average 100 Price', linewidth = 1)
plt.plot(ema200, label = f'{stock} Exp. Moving Average 200 Price', linewidth = 1)
plt.xlabel('Date')
plt.ylabel('High (USD)')
plt.legend()
plt.show()
```
